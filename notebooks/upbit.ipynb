```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pyupbit
print(pyupbit.Upbit)
```

```python
tickers = pyupbit.get_tickers(fiat="KRW")
print(tickers)
```

```python
price = pyupbit.get_current_price("KRW-BTC")
print(price)
```

```python
df = pyupbit.get_ohlcv("KRW-BTC")
df
```

```python
df = pyupbit.get_ohlcv("KRW-BTC", interval="minute1", count=1000)
df
```

```python
fig = plt.figure(figsize=(10,4))
sns.lineplot(data=df, x=df.index, y='close')
plt.show()
```

```python
df['close_diff'] = df['close'].diff().dropna()
fig = plt.figure(figsize=(10,4))
sns.lineplot(data=df, x=df.index, y='close_diff')
plt.show()
```

```python
df['close_diff2'] = df['close_diff'].diff().dropna()
fig = plt.figure(figsize=(10,4))
sns.lineplot(data=df, x=df.index, y='close_diff2')
plt.show()
```
